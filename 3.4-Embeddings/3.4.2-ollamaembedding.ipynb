ollama

ollama supports embedding models, making it possible to build retrival augmented generation(RAG) applications that combine text prompts with existing documnets or other data.

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings=(OllamaEmbeddings(model="gemma:2b"))## by default it uses llama2
#but i download in cmd gemma:2b
embeddings
```

```python
r1=embeddings.embed_documents(
    [
        "Alpha is the first letter of greek alphabet",
        "Beta is the second letter of greek alphabet",
    ]
)
```

```python
len(r1[0])
```

```python
r1[1]
```

```python
embeddings.embed_query("what is the second letter of greek alphabet")
```

```python
#other embedding models
#https://ollama.com/blog/embedding-models
embeddings=OllamaEmbeddings(model="mxbai-embed-large")
text="This is test documnet"
query_result=embeddings.embed_query(text)
query_result
```
